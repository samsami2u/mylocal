```python
import random

values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

suits = {'Club', 'Spade', 'Heart', 'Diamond'}
ranks = {'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace'}

playing = True

class Card:
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        
        return f'{self.rank} of {self.suit}'
    
```

```python
class Deck:
    
    def __init__(self):
        
        self.deck = []
        
        for suit in suits:
            for rank in ranks:
                
                current_card = Card(suit, rank)
                self.deck.append(current_card)
                
    def shuffle_deck(self):
        random.shuffle(self.deck)
        
    def deal_one(self):
        return self.deck.pop()
    
    def __str__(self):
        
        for card in self.deck:
            print(card)
        return 'The list of cards the deck has are'
```

```python
class Chips:
    
    def __init__(self):
        
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
        
    def take_bets(self):
        
        while True:
            try:
                bet_amount = input('Please input your bet for the game')
                self.bet = int(bet_amount)
            except:
                print('Please enter the amount as integer')
            else:
                break
                
        # self.bet = int(bet_amount)
```

```python
class gameService:
    
    def __init__(self):
        pass
    
    def hit_player(self, hand, deck):
        
        new_card = deck.deal_one()
        
        # if the deal card is an ace update the player hand accordinly
        if new_card.rank == 'Ace':
            hand.aces += 1 # increase ace count
            
            # if the current hand value plus the ace value as 11 busts then consider ace value as 1 or else 11
            if hand.value + 11 > 21:
                hand.value += 1
            else:
                hand.value += 11
            
            # if the card is not an ace add the card value to the hand value
        else:
            hand.value += new_card.value
        
        hand.cards.append(new_card)
        
    def hit_dealer(self, hand, deck):
        
        new_card = deck.deal_one()
        
        # if the deal card is an ace update the player hand accordinly
        if new_card.rank == 'Ace':
            hand.aces += 1 # increase ace count
            
            # if the current hand value plus the ace value as 11 busts then consider ace value as 1 or else 11
            if hand.value + 11 > 17:
                hand.value += 1
            else:
                hand.value += 11
            
            # if the card is not an ace add the card value to the hand value
        else:
            hand.value += new_card.value
        
        hand.cards.append(new_card)
        
    def hit_or_stand(self, hand, deck):
        global playing
        
        while True:
            
            player_response = input('\nDo you want to take a hit or stand? H/S')
            
            if player_response == 'H':
                print('Hit')
            
            if player_response != 'H' and player_response != 'S':
                print(player_response)
                print('Please enter the input as either "H" or "S"')
            else:
                break
                
        if player_response == 'H':
            playing = True
            return True
        else:
            playing = False
            return False
            
    def show_some(self, player, dealer):
        print('\nPlayer cards are:')
        for card in player.cards:
            print(card)
            
        print('\nDealer cards are:')
        dealer_cards = dealer.cards
        for index, card in enumerate(dealer.cards):
            if index > 0:
                print(card)
    
    def show_all(self, player, dealer):
        print('\nPlayer cards are:')
        for card in player.cards:
            print(card)
            
        print('\nDealer cards are:')
        for card in dealer.cards:
            print(card)
            
    def player_busts(self):
        print('\nPlayer lost the game and busted!')
    
    def player_wins(self):
        print('\nPlayer Won!')
    
    def dealer_busts(self):
        print('\nDealer lost the game and busted!')
    
    def dealer_wins(self):
        print('\nDealer Won!')
    
    def push():
        pass
```
